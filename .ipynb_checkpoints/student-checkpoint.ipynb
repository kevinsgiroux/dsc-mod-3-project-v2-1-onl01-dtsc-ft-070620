# Introduction

As a lifelong musician, I am personally very interested in the problem of genre identification for music where the genre is unknown, using measurable data extracted from the waveform of a given audio sample. 

As a stepping stone toward the above project, I will use this notebook to teach a supervised classification model to differentiate between 10 different musical genres, based on 10 features retrieved from track metadata as stored in Spotify's API (Note:  some of these features can be derived from the track waveform, e.g. 'loudness', others of which cannot, e.g. 'popularity').

The primary goals of this work are (1) to develop an understanding for which features are most important to the musical genre classification problem and (2) to get a sense for whether genre classification based purely on an audio waveform or audio fingerprint might be possible.

## Categorize subgenres into larger macrogenres groupings for classification

```python
genres = df_gen.genres

freq = {}
```

```python
# delimit the list of genre classes for each artist to generate one long list of every genre 
# included in the dataset

listy = []
for i in genres:
    i = i[1:-1].split(sep=',')
    for item in i:
        listy.append(item)

listy = pd.DataFrame(listy)
listy = listy[listy[0] != '']
```

```python
# remove spaces from the fronts of the delimited genres

newlist = []
for i in listy[0]:
    if i[0] == ' ':
        newlist.append(i[1:])
    else:
        newlist.append(i)
```

```python
# add genres and their counts to a dictionary

for i in newlist:
    if (i in freq): 
        freq[i] += 1
    else: 
        freq[i] = 1
```

```python
len(freq)
```

```python
# sort dictionary by count to show frequency of every genre included in the dataset

{k: v for k, v in sorted(freq.items(), key=lambda item: item[1], reverse=True)}
```

```python
# create a genre dictionary that groups select sub-genres into larger macro-genre categories for classification

genredict = {'pop': ['pop', 'indie pop', 'new wave pop', 'electropop', 'art pop', 'europop'],
             'rock': ['rock'],
             'hiphop': ['hip hop', 'rap', 'urban contemporary'],
             'country': ['country', 'contemporary country', 'traditional folk', 'country road'],
             'blues': ['blues', 'electric blues', 'traditional blues', 'modern blues', 'acoustic blues', 
                     'chicago blues', 'soul blues', 'new orleans blues', 'blues rock'],
             'reggae': ['reggae', 'roots reggae', 'reggae fusion', 'modern reggae', 'east coast reggae', 
                      'early reggae', 'west coast reggae'],
             'metal': ['metal', 'alternative metal', 'metalcore', 'death metal', 'industrial metal', 
                       'melodic metalcore', 'glam metal', 'nu metal', 'funk metal', 'power metal', 
                       'progressive metal', 'thrash metal', 'groove metal', 'speed metal'],
             'jazz': ['jazz', 'cool jazz', 'vocal jazz', 'smooth jazz', 'jazz quartet', 'jazz trio', 
                    'contemporary jazz'],
             'disco': ['disco', 'funk', 'post-disco', 'disco house', 'nu disco', 'italian disco', 'deep disco'],
             'classical': ['classical', 'classical performance', 'opera', 'orchestral performance',
                         'early modern classical']}
```

## Data cleaning

### Drop rows where genre column is blank

```python
df_gen = df_gen.loc[(df_gen['genres'] != '[]')]
df_gen = df_gen.reset_index()
df_gen = df_gen.drop('index', axis=1)
```

### Reformat genre column so each row is a list of subgenres rather than one long string
- chop the last character off each row in genre
- split each row by comma
- chop the first 2 characters and last 1 character off each item after the split

```python
df_gen['genre-1'] = df_gen['genres'].map(lambda x: x[:-1])
df_gen['genre_split'] = df_gen['genre-1'].str.split(pat=',')
```

```python
genre_lists = []

for i in df_gen['genre_split']:
    clean = []
    for item in i:
        clean.append(item[2:-1])
    genre_lists.append(clean)
```

### Combine subgenres in list form with the original dataframe

```python
genre_lists = pd.Series(genre_lists)
df_genre = pd.concat([df_gen,genre_lists], axis=1)
```

```python
df_genre = df_genre.drop(['genres', 'genre-1', 'genre_split'], axis=1)
df_genre.rename(columns={0:'genre'}, inplace=True)
```

### Use genredict mappings to go line by line through df_genre and classify each artist (aka each row) into one of the predetermined macrogenres categories based on that artist's list of associated subgenres

```python
# Iterate through genredict for each macrogenre, and check the listed sub-genres for every row in the df
# Create a dataframe with a column for each macrogenre and the same indexing as the original dataset
# If a given row in the original data has one of the sub-genres contained within a macrogenre in genredict,
#     flag that row in the appropriate column of the generated macrogenre df

macrogenredf = None

for n, key in enumerate(genredict):

    genlist = []
    for i in df_genre.genre:
        
        for item in i:
            
            subgenlist = []
            for g in genredict.get(key):
                
                if g == item:
                    subgenlist.append(g)
                else:
                    continue
        genlist.append(str(subgenlist))
    
    if n == 0:
        macrogenredf = pd.Series(genlist, name=key)
    else:
        macrogenredf = pd.concat([macrogenredf, pd.Series(genlist, name=key)], axis=1)
    
```

```python
for i in macrogenredf.columns:
    macrogenredf[i].loc[macrogenredf[i] == '[]'] = 0
```

# Preprocessing

The goal of this section is to establish 4 versions of the test and training data that I can use on each of the various classification models:

1. Unmodified, as pulled from the original dataset
2. Scaled, so all feature values fall between 0 and 1
3. Scaled with only 7 features, (same as 2, but with the 3 lowest-variance features removed)
4. Preprocessed to meet the assumptions behind logistic (i.e. linear) regression.
5. Corrected for class-weight imbalances using SMOTE (scaled, 10 predictors)

### Transform non-normally distributed features, where possible
- 'liveness' and 'speechiness' should be log- or sqrt-transformed or  to correct for right-hand skew
- 'tempo' and 'danceability' appear fairly normally distrubuted
- 'valence' has fat tails but otherwise normal
- 'energy' is slightly bi-modal but otherwise normal
- 'popularity' has fat tails and some left-hand skew -- not completely sure how to correct for this, so leaving it as it is for now

```python
# liveness and speechiness should be log- or sqrt-transformed or  to correct for right-hand skew
```

```python
live_tr = xtrain_scaled['liveness']
speech_tr = xtrain_scaled['speechiness']

live_te = xtest_scaled['liveness']
speech_te = xtest_scaled['speechiness']

live_tr = np.log(live_tr) / -5 
speech_tr = np.sqrt((round(np.log2(speech_tr), 5) + 6)) / 2.25
                    
live_te = np.log(live_te) / -5 
speech_te = np.sqrt((round(np.log2(speech_te), 5) + 6)) / 2.25
```

```python
sns.distplot(xtrain_scaled['liveness'], label='Non-transformed data')
sns.distplot(live_tr, label='Transformed data')
plt.legend()
```

```python
sns.distplot(xtrain_scaled['speechiness'], label='Non-transformed data')
sns.distplot(speech_tr, label='Transformed data')
plt.legend()
```

```python
xtrain_scalog = xtrain_scalog.drop(columns=['liveness', 'speechiness'], axis=1)
```

```python
xtrain_scalog = pd.concat([xtrain_scalog, live_tr, speech_tr], axis=1)
```

# Initial model selection

```python
def basescore(model, xtrainlist, xtestlist, ytrainlist):
    scores = []
    for n, x in enumerate(xtrainlist):
        model.fit(x, ytrainlist[n])
        cvscore = np.mean(cross_val_score(model, xtestlist[n], y_test, cv=5))
        scores.append(cvscore)
    return scores
```

### Parameter tuning

#### Optimal regularization parameter: 100
Test score: .617

```python
cs = [0.001, 0.01, 0.1, 1, 10, 100, 1000, 10000]
results = []
train_results = []
test_results = []

for c in cs:
    log = LogisticRegression(C=c, max_iter=10000, fit_intercept=True, random_state=29)
    log.fit(xtrain_scaled, y_train)
    train_pred = log.predict(xtrain_scaled)
    test_pred = log.predict(xtest_scaled)
    trscore = np.mean(cross_val_score(log, xtrain_scaled, y_train, cv=5))
    tescore = np.mean(cross_val_score(log, xtest_scaled, y_test, cv=5))
    results.append((c, round(trscore, 3), round(tescore, 3)))
    train_results.append(trscore)
    test_results.append(tescore)

x = np.linspace(1, 8, 8)
plt.figure(figsize=(12,6))
plt.plot(x, train_results, 'b', label='Train Score')
plt.plot(x, test_results, 'r', label='Test Score')
plt.ylabel('Score')
plt.xlabel('Regularization parameter')
plt.xticks(ticks=x, labels=cs)
plt.legend()
plt.show()
```

```python
results
```

#### Optimal max_iterations: 300 +
Above a value of 300, the max_iter parameter does not affect the model's classification score for either the training or testing datasets.  Below a value of 300, the gradient descent function does not successfully converge.

```python
maxiter = list(range(300, 1500, 100))
results = []
train_results = []
test_results = []

for m in maxiter:
    log = LogisticRegression(C=10, max_iter=m, fit_intercept=True, random_state=29)
    log.fit(xtrain_scaled, y_train)
    train_pred = log.predict(xtrain_scaled)
    test_pred = log.predict(xtest_scaled)
    trscore = log.score(xtrain_scaled, y_train)
    tescore = log.score(xtest_scaled, y_test)
    results.append((m, round(trscore, 3), round(tescore, 3)))
    train_results.append(trscore)
    test_results.append(tescore)
    
plt.figure(figsize=(12,6))
plt.plot(maxiter, train_results, 'b', label='Train Score')
plt.plot(maxiter, test_results, 'r', label='Test Score')
plt.ylabel('Score')
plt.xlabel('Regularization parameter')
plt.legend()
plt.show()
```

## Gradient Boosted Trees

```python
gbc = GradientBoostingClassifier(learning_rate=.25, 
                                 n_estimators=300,
                                 min_samples_leaf=8,
                                 max_depth=3,
                                 min_samples_split=3,
                                 max_features=2,
                                 random_state=29)
gbc.fit(xtrain_smote, ytrain_smote)
gbc.score(xtest_scaled, y_test)
```

### Feature importances

```python
featimps_gbc = gbc.feature_importances_
x = range(len(featimps_gbc))

plt.figure(figsize=(6,4))
plt.bar(x, featimps_gbc)
plt.xticks(ticks=x, labels=coeflist_10, rotation=45, ha='right');
plt.title('Feature Importances: Gradient Boosted Trees')
```

Conclusions:
- Overall model F1 score:  66.2%


- 5 of top 6 most important features are derived from the audio fingerprint / waveform of a track
    1. Acousticness
    2. Energy
    3. Danceability
    4. Speechiness
    5. Valence
    6. Popularity


- Popularity is the only feature in the top 6 not deriveable from a waveform


- Least important to genre prediction are loudness, liveness, instrumentalness, and tempo 
- This result makes sense, as tracks within each genre category could easily vary widely along all four of these parameters, depending, for example, on 
    1. How each individual track was mastered (loudness), 
    2. Whether the track is a live or studio recording (liveness), 
    3. If the vocal stems of a track were removed to create an instrumental mix (instrumentalness).  
    4. Similarly, tempo can vary significantly within a given genre category.

# Conclusions and Recommendations


## Conclusions

1. Model accuracy
    - Gradient Boosted Trees F1 score:  66.2%
    - Random Forest F1 score:  64.1%


2. 5 of top 6 most important features are derived from the audio fingerprint / waveform of a track
    - Acousticness
    - Energy
    - Danceability
    - Speechiness
    - Valence


3. Popularity is the only feature in the top 6 not deriveable from a waveform


4. Least important to genre prediction are loudness, liveness, instrumentalness, and tempo
    - This result makes sense, as tracks within each genre category could easily vary widely along all four of these parameters, depending, for example, on
        - How each individual track was mastered (loudness),
        - Whether the track is a live or studio recording (liveness),
        - If the vocal stems of a track were removed to create an instrumental mix (instrumentalness).
        - Similarly, tempo can vary significantly within a given genre category


5. Precision and recall scores were both about equal for both models, with the same strenghths and weaknesses as far as which genres the models were best and worst at classifying.
    - The models were both best at identifying Classical, Metal, and Hip Hop
    - The models were both bad at identifying Blues and Disco
    - These similar results across models could be due to the fact that both models rely on tree-based algorithms

## Business recommendations
1. Based on the above conclusions, it appears data derived from an audio waveform IS useful as a predictor for the track’s genre; thus we should greenlight phase 2 of app development.
        
2. Though genre prediction via track waveform seems feasible, prediction accuracy may be improved with the addition of non-audio metadata features.  This should be considered by both the business and technology teams.

3. The models in their current form tend to work best on extreme-sounding genres (e.g. Classical, Metal, and Hip Hop), but tends to confuse similar sounding genres (e.g. Blues, Jazz, and Rock).  Thus the final product should have a feature that offers 2nd and 3rd choice genre classification predictions when a given prediction is below a certain threshold of certainty.

## Next steps / Future work
1. Expand the dataset to include additional unique examples of Blues and Disco tracks, which were the most difficult for the models to classify correctly.
2. Experiment with building a series of models that are trained only on the top 6 most important features identified in this phase of the project.
3. Kick off an in-depth genre mapping exercise which will inform the training of future, more sophisticated versions of these models.
